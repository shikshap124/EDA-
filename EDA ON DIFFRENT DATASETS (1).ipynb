```python


# Drop rows with nulls (if any)
df.dropna(inplace=True)

# Check for duplicates
print(f"Duplicate rows: {df.duplicated().sum()}")

# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
#Q2 What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=30, kde=True, color='purple')
plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.show()
```

```python
3.# Popularity vs Duration (Scatter Plot)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Duration (ms)', y='Popularity', hue='Popularity', palette='viridis')
plt.title("Popularity vs Duration")
plt.xlabel("Duration (ms)")
plt.ylabel("Popularity")
plt.show()
```

```python
# Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using 
a countplo
plt.figure(figsize=(14, 6))
top_artists = df['Artist'].value_counts().head(10)
sns.countplot(data=df[df['Artist'].isin(top_artists.index)], y='Artist', order=top_artists.index, palette='mako')
plt.title("Top Artists by Number of Tracks")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.show()
```

```python
#What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each
least_popular = df.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]
print(least_popular)
```

```python
# Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and 
#display the average popularity for each artist
 
 
top_artists_by_track = df['Artist'].value_counts().head(5).index
top_artist_avg_pop = df[df['Artist'].isin(top_artists_by_track)].groupby('Artist')['Popularity'].mean().sort_values(ascending=False)
print(top_artist_avg_pop)
```

```python
#For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist
top_tracks = df[df['Artist'].isin(top_artists_by_track)]
most_popular_tracks = top_tracks.sort_values('Popularity', ascending=False).drop_duplicates('Artist')[['Artist', 'Track Name', 'Popularity']]
print(most_popular_tracks)
```

```python
# Visualize relationships between multiple numerical variables simultaneously using a pair plot
sns.pairplot(df[['Popularity', 'Duration (ms)']])
plt.show()
```

```python
# Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or 
#violin plot
plt.figure(figsize=(14, 6))
sns.violinplot(data=df[df['Artist'].isin(top_artists.index)], x='Artist', y='Duration (ms)', palette='coolwarm')
plt.title("Track Duration Distribution Across Artists")
plt.xticks(rotation=45)
plt.show()
```

```python
# How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a 
violin plot
plt.figure(figsize=(14, 6))
sns.swarmplot(data=df[df['Artist'].isin(top_artists.index)], x='Artist', y='Popularity', palette='Spectral')
plt.title("Popularity Distribution by Artist")
plt.xticks(rotation=45)
plt.show()
```
